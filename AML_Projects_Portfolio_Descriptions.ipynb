Best Prediction Model Description:

To develop the prediction models, I first established the variables as feature and target variables. I then split the data into training and test sets. I then further processed the data.

The final epoch returned an accuracy of .68 and loss of .79. 






### Assignment #2
 Link to notebook - https://github.com/Paarthaenon/Adv.-ML-Public-Repo/blob/master/Advanced_ML_HW_2.ipynb

Description - The primary goal of assignment #2 was to predict if an image of a brain had a tumor or not. To train the classification model, I used several brain images, some with tumors and some without. A convolutional neural network was used to predict. 

Use - By having a model that can predict if a brain image has a tumor, healthcare workers could better diagnose patients by reducing human error. 

Data shape - (253 rows x 2 columns)
              A total of 253 pictures were used to train the model. The pictures were designated with either "no" it doesn't have a tumor or "yes" it does. 

Best Model Description: The model which returned the greatest test set accuracy is an inception model. The inception model utilizes transfer learning, which takes pre-trained weights from another model which performs well. By using the pre-trained weights, my own model does not need to go through the lengthy process of forward and backward propogation to determine the ideal weights the neural network should use. In this case, the inception model utilizes pre-trained weights from imagenet. The single dense layer of 64 nodes uses relu activation and has dropout of .3. The output layer has two outputs (yes or no tumor) and sigmoid activation. I compiled the model with sgd optimizer, binary crossentrioy loss (since there's two outputs) and the accuracy metrics. The fit model runs through 10 epochs. The returned test set accuracy is .88, with a loss of .34.
